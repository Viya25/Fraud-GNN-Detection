```python
import pandas as pd

df = pd.read_csv(r"C:\Users\Asus\Documents\prj\Fraud-GNN-Detection\data\creditcard.csv")
```

```python
from sklearn.preprocessing import StandardScaler

# split features & label
X = df.drop("Class", axis=1)
y = df["Class"]

# scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Scaled shape:", X_scaled.shape)
```

```python
import torch

x = torch.tensor(X_scaled, dtype=torch.float)
y = torch.tensor(y.values, dtype=torch.long)

print("Tensor shape:", x.shape)
```

```python
from sklearn.neighbors import NearestNeighbors
from torch_geometric.data import Data
import torch

# build KNN graph
knn = NearestNeighbors(n_neighbors=5)
knn.fit(X_scaled)

# get sparse matrix (IMPORTANT)
edges = knn.kneighbors_graph(X_scaled)

# convert directly to edge_index (NO toarray)
edge_index = torch.tensor(
    [edges.nonzero()[0], edges.nonzero()[1]],
    dtype=torch.long
)

# create graph
data = Data(x=x, edge_index=edge_index, y=y)

print(data)
```
